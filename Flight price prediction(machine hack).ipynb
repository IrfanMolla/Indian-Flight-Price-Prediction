# making new dataframe,stops,  from  Route

```python
def routeTostops(dataset):    
    lst1=list(dataset.loc[:,'Route'])

    for i in range(len(lst1)):
        if type(lst1[i])!=str:
            print(i)
        else:
            lst1[i]=lst1[i].split(' → ')

    list_of_airports=set()

    for i in range(len(lst1)):
        for j in range(len(lst1[i])):
            list_of_airports.add(lst1[i][j])

    list_of_airports=list(list_of_airports)

    for i in range(len(lst1)):
        for j in range(len(lst1[i])):
            lst1[i][j]=int(list_of_airports.index(lst1[i][j])+1)

    for i in range(len(lst1)):
        if len(lst1[i])!=6:
            if len(lst1[i])==2:
                for j in range(0,4):
                    lst1[i].append(0)
            elif len(lst1[i])==3:
                for j in range(0,3):
                    lst1[i].append(0)
            elif len(lst1[i])==4:
                for j in range(0,2):
                    lst1[i].append(0)
            elif len(lst1[i])==5:
                for j in range(0,1):
                    lst1[i].append(0)

    stops=pd.DataFrame(lst1,columns=[1,2,3,4,5,6])
    return stops
```

```python
def airlines(dataset):
    lst2=set()
    for i in range(len(dataset)):
        lst2.add(dataset.loc[i,'Airline'])
    lst2=list(lst2)
    list_of_airlines=list()
    for i in range(len(dataset)):
        list_of_airlines.append(dataset.iloc[i,0])

    for i in range(len(list_of_airlines)):
        list_of_airlines[i]=int(lst2.index(list_of_airlines[i]))

    airline=pd.DataFrame(list_of_airlines,columns=['Airline'])
    return airline
```

```python
def dofj(dataset):
    lst3=list()
    for i in range(len(dataset)):
        lst3.append(dataset.loc[i,'Date_of_Journey'])

    for i in range(len(lst3)):
        lst3[i]=lst3[i].split('/')

    for i in range(len(lst3)):
        for j in range(len(lst3[i])):
            if type(lst3[i][j])==str:
                lst3[i][j]=int(lst3[i][j])

    Date_of_Journey=pd.DataFrame(lst3,columns=['Date','Month','Year'])
    return Date_of_Journey
```

# new source and destination dataframe

```python
def getsource(dataset):
    list_of_source=list(set(dataset.loc[:,'Source']))
    source=list(dataset.loc[:,'Source'])
    for i in range(len(source)):
        source[i]=int(list_of_source.index(source[i]))
    source=pd.DataFrame(source,columns=['source'])
    return source
```

```python
def getdestination(dataset):
    list_of_destination=list(set(dataset.loc[:,'Destination']))
    destination=list(dataset.loc[:,'Destination'])
    for i in range(len(destination)):
        destination[i]=int(list_of_destination.index(destination[i]))
    destination=pd.DataFrame(destination,columns=['Destination'])
    return destination
```
